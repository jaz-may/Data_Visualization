```python
import networkx as nx
import matplotlib
matplotlib.use("TkAgg")
import matplotlib . pyplot as plt
G = nx.Graph() # Undirected Graph
G.add_edge("A", "B")
G.add_edge("A", "C")
G.add_edge("B", "C")
G.add_edge("C", "D")
G.add_edge("B", "D")
pos = nx.circular_layout(G)
plt.figure(figsize = (5, 5))
nx.draw(G, pos, with_labels = True,
        node_color ="lightblue",
        node_size = 2500
        )
plt.axis ("off")
plt.show()
```

```python
# 有向图
import networkx as nx
import matplotlib
matplotlib.use("TkAgg")
import matplotlib . pyplot as plt
G = nx.DiGraph() # Undirected Graph
G.add_edge("A", "B")
G.add_edge("A", "C")
G.add_edge("B", "C")
G.add_edge("C", "D")
G.add_edge("B", "D")
pos = nx.circular_layout(G)
plt.figure(figsize = (5, 5))
nx.draw(G, pos, with_labels = True,
        node_color ="lightblue",
        node_size = 2500
        )
plt.axis ("off")
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
adj_matrix = np.array([
                        [0,1,1],
                        [0,0,1],
                        [1,1,0]
                      ]
                     )
G = nx.from_numpy_array(adj_matrix, create_using = nx.DiGraph())
mapping = {0: "A", 1: "B", 2: "C"}
G = nx.relabel_nodes(G, mapping)
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels = True, node_color = "orange", node_size = 2500 , arrowsize = 60)
plt.title("Graph from Adjacency Matrix")
plt.axis("off")
plt.show()
```

```python
import networkx as nx
import matplotlib . pyplot as plt
# 1. Create a weighted graph
G = nx.Graph()
G.add_edge("A", "B", weight = 1)
G.add_edge("A", "C", weight = 3)
G.add_edge("B", "C", weight = 5)
G.add_edge("C", "D", weight = 2)
# 2. Layout
pos = nx.spring_layout(G, seed = 42)
# 3. Edge width ~ weight
weights = [G[u][v]["weight"] for u, v in G.edges()]
edge_widths = [w * 1.5 for w in weights]
# 4. Node size ~ degree
degrees = dict(G.degree())
node_sizes = [degrees[n] * 500 for n in G.nodes()]
# 5. Draw graph
plt.figure(figsize =(5 ,5))
nx.draw(G, pos, with_labels = True, width = edge_widths, node_size = node_sizes, node_color = "lightblue", edge_color = "grey")
plt.title("Graph Visualization")
plt.axis("off")
plt.show()
```

```python
fig , ax = plt . subplots ( figsize =(5 ,5) )
nx. draw (G, pos , with_labels =True ,
node_size = node_sizes ,
node_color ="lightblue",
edge_color = weights ,
edge_cmap = plt .cm. viridis ,
ax=ax
) 
# colorbar
sm = plt .cm. ScalarMappable ( cmap =plt.cm. viridis , norm = plt . Normalize ( vmin = min ( weights ),
vmax = max ( weights )))
sm. set_array ([])
fig . colorbar (sm , ax=ax , label ="Edge Weight")
plt . title ("Graph Visualization")
plt . axis ("off")
plt.show()
```
