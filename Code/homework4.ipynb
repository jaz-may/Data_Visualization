```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
# task 1
with open("../作业/answers.txt", "r") as f:
    lines = [line.strip() for line in f.readlines() if line.strip()]

replace_map = {"A": 1, "B": 2, "C": 3, "D": 4}
matrix = []
for line in lines:
    num_line = [replace_map[char] for char in line]
    matrix.append(num_line)

# 构造X 
X = np.array(matrix)
X
```

```python
# task 2
sample_num , feature_num = X.shape
print(f"number of samoles:{sample_num}\nnumber of features:{feature_num}")
```

```python
# task 3
pca = PCA(n_components = 2)
data = pca.fit_transform(X)
```

```python
# task 4: reconstruction
data_reconstruct = pca.inverse_transform(data)
```

```python
# task 5
# 转回字符
data_rounded = np.clip(np.round(data_reconstruct), a_min = 1, a_max = 4)   # 转回整数
res = []
reconstruction_map = {1:"A", 2:"B", 3:"C", 4:"D"}
for row in data_rounded:
    char_row = [reconstruction_map[i] for i in row]
    res.append(char_row)
print(res)
```

```python
# task 6: Accuracy
total = 20 * 25
bias = 0
for i in range(len(res)):
    for j in range(len(res[0])):
        if res[i][j] != lines[i][j]:
            bias += 1
print(f"Accuracy:{(total - bias) / total}")
```
